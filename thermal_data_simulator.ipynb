```python
!pip install paho-mqtt
import paho.mqtt.client as mqtt
import time
import json
import random

# --- CONFIGURATION ---
THINGSBOARD_HOST = 'demo.thingsboard.io'
ACCESS_TOKEN = 'b4ee7h88l3sp0h6dycmo'
INTERVAL = 5

# --- GAUSSIAN MODEL PARAMETERS ---
TEMP_MEAN = 42.5    # The average temperature in Celsius
TEMP_STD_DEV = 0.8  # The "wiggle" (Standard Deviation)
POWER_MEAN = 4.2    # The average power in Watts
POWER_STD_DEV = 0.3 # Small fluctuations in electrical draw

# --- VIRTUAL SENSOR LOGIC ---
def get_simulated_data():
    # Generate Gaussian "Stable" data
    temperature = random.gauss(TEMP_MEAN, TEMP_STD_DEV)
    power = random.gauss(POWER_MEAN, POWER_STD_DEV)

    # 10% chance to simulate a "Fault Injection" (Overheat)
    if random.random() < 0.10:
        print("!!! ANOMALY DETECTED: SIMULATING OVERHEAT !!!")
        temperature = random.uniform(55.0, 62.0)

    return {
        "temperature": round(temperature, 2),
        "power": round(power, 2)
    }

# --- MQTT SETUP ---
client = mqtt.Client()
client.username_pw_set(ACCESS_TOKEN)

try:
    print(f"Connecting to {THINGSBOARD_HOST}...")
    client.connect(THINGSBOARD_HOST, 1883, 60)
    client.loop_start()

    while True:
        data = get_simulated_data()
        payload = json.dumps(data)

        print(f"Streaming Telemetry [Normal Dist]: {payload}")
        client.publish('v1/devices/me/telemetry', payload, qos=1)

        time.sleep(INTERVAL)

except KeyboardInterrupt:
    print("Simulation stopped.")
finally:
    client.loop_stop()
    client.disconnect()
```
